```python
import os
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
df = pd.read_csv('filtered_x1x2_wo_copies.tsv', sep='\t', index_col=0)
```

```python
df
```

```python
dist = pd.DataFrame(columns=df.index, index=df.index)
```

```python
samples = df.index
```

```python
for i in tqdm(samples):
    tmp_dict = {}
    for j in samples:
        temp = np.array(df.loc[i]) - np.array(df.loc[j])
        tmp_dict[j] = np.sqrt(np.dot(temp.T, temp))
    dist.loc[i] = tmp_dict
```

```python
dist
```

```python
dist.to_csv('distance_matrix.tsv', sep='\t', index=True)
```

```python
phenotypes = pd.read_csv('phenotypes.tsv', sep='\t', index_col=0)
```

```python
phenotypes.head()
```

```python
dist_dict = {}
for sample in samples:
    dist_dict[sample] = dist[sample].sort_values()[1:].index
```

```python
for year in phenotypes.columns:
    for sample in phenotypes[year].index:
        if pd.isna(phenotypes.loc[sample, year]):
            non_na_neighbours = []
            i = 0
            while len(non_na_neighbours) < 5:
                if not pd.isna(phenotypes.loc[dist_dict[sample][i], year]):
                    non_na_neighbours.append(dist_dict[sample][i])
                i += 1
            phenotypes.loc[sample, year] = round(np.average([phenotypes.loc[_, year] for _ in non_na_neighbours]))
```

```python
phenotypes
```

```python
phenotypes.to_csv('phenotypes_adjusted.tsv', sep='\t', index=True)
```
