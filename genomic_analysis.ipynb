```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
vcf = pd.read_csv('genotypes.vcf', sep='\t', skiprows=1216)
```

```python
vcf.rename(columns = {'#CHROM':'CHROM'}, inplace=True)
```

```python
vcf
```

```python
# num of samples in VCF file
samples = vcf.columns[9:]
len(samples)
```

```python
# Double check, num samples in phenotypes
pd.read_csv('phenotypes.tsv', sep='\t').shape[0]
```

```python
mut_dict = {}
for sample in tqdm(samples):
    for index in vcf[sample].index:
        GT, PL, _ = vcf[sample][index].split(':')
        if '.' not in GT:
            variants = [vcf.loc[index]['REF']] + vcf.loc[index]['ALT'].split(',')
            alleles = map(int, GT.split('/'))
            for allele in alleles:
                if allele != 0:
                    chrom, pos = vcf.loc[index]['CHROM'], vcf.loc[index]['POS']
                    ref, alt = variants[0], variants[allele]
                    subst = f'{chrom}:{pos}:{ref}-{alt}'
                    if GT[0] == GT[-1]:
                        try:
                            mut_dict[subst][sample] = 2
                        except KeyError:
                            mut_dict[subst] = {}
                            mut_dict[subst][sample] = 2
                    else:
                        try:
                            mut_dict[subst][sample] = 1
                        except KeyError:
                            mut_dict[subst] = {}
                            mut_dict[subst][sample] = 1
```

```python
subst_df = pd.DataFrame(columns=mut_dict.keys(), index=samples)
```

```python
subst_df = subst_df.from_dict(mut_dict).fillna(0)
```

```python
subst_df.to_csv('unfiltered_1_2.tsv', sep='\t', index=True)
```

```python
gff = pd.read_csv('Glycine_max.Glycine_max_v2.1.60.chr.gff3', sep='\t', comment = '#', header = None)
gff.columns = ['seqname', 'source', 'feature', 'start', 'end', 'score', 'strand', 'frame', 'attribute']
```

```python
gff.seqname.unique() #Тут надо уточнить, что в аннотации нет мутаций не из аннотированных хромосом, поэтому их тоже дропну
```

```python
gff = gff[gff.attribute.str.contains('protein_coding')][gff.feature == 'gene'].reset_index()
gff.drop(columns=['index'])
```

```python
gff.shape #Уменьшили всю аннотацию только до кодирующих генов, чтоб считалось быстрее
```

```python
filtered_columns = []
for column in tqdm(subst_df.columns):
    try:    
        chrom, pos = map(int, column.split(':')[:2])
    except ValueError:
        continue
    df = gff[gff.start < pos][gff.end > pos]
    if df.shape[0] > 0:
        filtered_columns.append(column)
```

```python
len(filtered_columns)
```

```python
subst_df[filtered_columns] #отфильтровали пару тысяч некодирующих/нехромосомных мутаций
```

```python
subst_df[filtered_columns].to_csv('filtered_x1x2.tsv', sep='\t', index=True)
```

```python
same_filter = []
for column in tqdm(filtered_columns):
    if not (all([item == 0.0 for item in subst_df[column].values])\
            or all([item == 1.0 for item in subst_df[column].values])):
        same_filter.append(column)
```

```python
len(same_filter)
```

```python
df = subst_df[same_filter].copy()
```

```python
values_used = []
columns_to_keep = []
for column in tqdm(same_filter):
    if tuple(df[column].values) not in values_used:
        columns_to_keep.append(column)
        values_used.append(tuple(df[column].values))
```
